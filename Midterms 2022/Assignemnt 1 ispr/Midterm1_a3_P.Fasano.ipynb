#### Section 1: The Code

The libraries used to write the code for the assignment 3 are the following:

```python
# for data transformation
import numpy as np
# for visualizing the data
import matplotlib.pyplot as plt
# for opening the media file
import scipy.io.wavfile as wavfile
from scipy import signal
# for reading the files in a path
import os
```

A path to a folder that contains all the .wav files was defined and all the names of the files within it were read.

```python
path = './wavToPlot/'
arrayNames = os.listdir(path)
```

To make the plot more understandable the title of the spectrogram and the labels for the x and y axis of the plot  were written.  

```python
plt.title('Spectrogram of a wav file with ' + arrayNames[0].split(".")[0] +' note') 
plt.xlabel('Time')
plt.ylabel('Frequency')
```

The wav file was read using the scipy library that gave back the sampling frequency and the audio written in a 2d array.  
The audio was subsequently re-elaborated (aud in the code below) to make it mono, if it wasn't already, and to trim it to a standard length.    

```python
Fs, aud = wavfile.read(path+arrayNames[0])
# trim the first 2 second
first = aud[:int(Fs*2)] 
```

The following step was to use the sampled frequencies and the audio to plot the spectrogram using the signal library from scipy and using the spectrogram function.    
The spectrogram function returns an array of frequencies, an array of time and the spectrogram it self which is a matrix of float values.

```python
frequencies, times, spectrogram = signal.spectrogram(first, Fs)
```

The final step for plotting the spectrogram was to use the matplotlib.pyplot library to plot and save a picture of the spectrogram: 

```python
plt.pcolormesh(times, frequencies, 10*np.log10(spectrogram), shading='gouraud')
plt.savefig(arrayNames[0].split(".")[0]+'.png')
```

The code shown above, has to be repeted for each wav file in the folder that corresponds to the variable path:

```python
for fi in arrayNames:
        
        plt.title('Spectrogram of a wav file with ' + fi.split(".")[0] +' note') 
        plt.xlabel('Time')
        plt.ylabel('Frequency')
        
        Fs, aud = wavfile.read(path+fi)
        # trim the first 2 second
        first = aud[:int(Fs*2)] 

        frequencies, times, spectrogram = signal.spectrogram(first, Fs)
        plt.pcolormesh(times, frequencies, 10*np.log10(spectrogram), shading='gouraud')      
        plt.savefig(fi.split(".")[0]+'.png')
```
